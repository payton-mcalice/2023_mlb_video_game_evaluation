```python
import pandas as pd
import math
pd.set_option('display.max_columns', None)
```

```python
def pitcher_stats(year):
    ts = pd.read_csv(f"Top Speed/{year}_Top_Speed.csv").set_index("Unnamed: 0")
    ctrl = pd.read_csv(f"Control/{year}_Control.csv").set_index("Unnamed: 0")
    stam = pd.read_csv(f"Stamina/{year}_Stamina.csv").set_index("Unnamed: 0")
    brk = pd.read_csv(f"Breaking Balls/{year}_BreakingBall.csv").set_index("Unnamed: 0")
    ability = pd.read_csv(f"Abilities/{year}_Pitcher_Ability.csv").set_index('player_id')

    merge = (ts.merge(ctrl, how='inner', left_index=True, right_index = True)
            .merge(stam, how='inner', left_index=True, right_index = True)
            .merge(brk, how='inner', left_index=True, right_index = True))
    merge = merge.loc[:,~merge.columns.duplicated()].copy()
    merge = merge.drop_duplicates()
    
    columns = ['PLAYERNAME_x', 'Top Speed', 'Control', 'Stamina', 'Slider', 'Change', 'Curve', 'Sinker', 'Cutter', 'Splitter']
    merge = merge[columns]
    return merge.merge(ability, how='left', left_index=True, right_index = True)

pitchers = pitcher_stats(2022)
pitchers.head()
```

```python
pitchers.to_csv("Pitchers_Data.csv")
```
